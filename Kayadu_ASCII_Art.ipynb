# ASCII Art - Kayadu Lohar

```python
from google.colab import files
from PIL import Image

uploaded = files.upload()

# Get file name
filename = list(uploaded.keys())[0]

# Opening image
img = Image.open(filename)
# Convert to RGB to avoid errors with png transparency
img = img.convert("RGB")

# Getting original dimensions
width, height = img.size
```

```python
# Step 2: Width of image to be drawn
w_new = int(input("Enter width: "))

# Calculate new height
# Aspect ratio correction (0.55) because terminal characters are tall
h_new = int(height * (w_new / width) * 0.55)

print(f"Resizing to {w_new}x{h_new}...")
```

```python
# Step 3 & 4: Processing and Converting
# Define characters from dark to light
chars = "@%#*+=-:. "

scale = (len(chars) - 1) / 255

final_art = ""

# (height)
for y in range(h_new):
    # (width)
    for x in range(w_new):

        src_x = int(x * (width / w_new))
        src_y = int(y * (height / h_new))

        # pixel color (R, G, B)
        r, g, b = img.getpixel((src_x, src_y))

# Calculate (average brightness)
        gray = int(0.2989 * r + 0.5870 * g + 0.1140 * b)

        index = int(gray * scale)

        final_art += chars[index]

    final_art += "\n"
```

```python
# Step 5: Printing the result
print("\nThe ASCII Art:\n")
print(final_art)
```
